```python
## toss win impact on match result
toss = matches.groupby(['season','toss_winner']).winner.value_counts().reset_index(name='count')
```

```python
toss.head()
```

```python
toss['result']= np.where(toss.toss_winner==toss.winner,'won','lost')
toss_r = toss.groupby(['season', 'toss_winner','result'])['count'].sum().reset_index()
```

```python
toss_r.head()
```

```python
##  win or lost matches in seasons
for x in range (2008, 2017, 1):
    toss_r_x = toss_r[toss_r['season']==x]
    plot = sns.barplot(x='toss_winner', y = 'count', hue = 'result', data = toss_r_x)
    plot.set_title('Matches won/lost by toss winning teams \nSeason' +str(x))
    plot.set_xticklabels(toss_r_x['toss_winner'], rotation =30)
    plt.show()
```
